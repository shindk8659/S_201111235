```python
import requests
import urllib

from scrapy.selector import Selector

target=open('son.txt','w')

url = 'https://www.premierleague.com/players/4999/Son-Heung-Min/stats?se=54'
r=requests.get(url)
_html=r.text

stats=["Appearances","Goals","Wins","Losses"]
cal=[None]*4

nodes=Selector(text=_html).xpath("//*[@class='playerDetails']")
for node in nodes:
    number=node.css("div[class='number']::text").extract()
    name=node.css("h1>div::text").extract()
    if number and name:
        number=number[0].strip()
        name=name[0].strip()
    target.write(number+'\n')
    target.write(name+'\n')
    print number,name
nodes=Selector(text=_html).xpath("//*[@class='topStat']")
i=0
for node in nodes:
    value = node.css("span[class='stat']>span::text").extract()
    
    if value and stats:
    
        value = value[0].strip()
    target.write(value+'\n')
     
    print stats[i], value
    cal[i]=value
    i=i+1
target.close()  
total=int(cal[2])+int(cal[3])
winper=(float(cal[2])/total)*100
print "출장횟수:",total
print "출장 횟수 대비 승률:",winper,"%"
```
